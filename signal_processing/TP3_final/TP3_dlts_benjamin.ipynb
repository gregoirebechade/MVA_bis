## Problématique

On cherche à identifier un type d'émetteur de communication à partir de l'observation d'un signal provenant de l'émetteur
de 2048 échantillons IQ (In Phase / Quadrature) ie le signal prend des valeurs complexes. On représente la partie
réelle et la partie imaginaire du signal par deux canaux réel d'un signal multivarié.

L'émetteur peut provenir de 6 catégories différentes.
Les paramètres différenciant les différentes catégories sont
- le type de modulation
- la présence ou non de séquences pilotes et le cas échéant la structure de trame pilotes / données
- le débit de la transmission

Les signaux se propagent en champs libre et sont enregistrés par une antenne. Le signal reçu est transposé en bande de base c'est à dire que si le signal est transmis autour d'une fréquence centrale f0, une première étape de traitement du signal à la réception recentre le signal autour de la fréquence 0.


Les différents signaux observés dans ce TP sont entachés de différentes erreurs caractéristiques de la propagation
électromagnétique comme :
- modification aléatoire de la phase du signal lors de la transmission
- imperfection de la transposition en bande de base qui laisse le signal transposé à une fréquence df0 << f0
- présence d'interférence entre les symboles transmis (dûes par exemple à plusieurs chemins de propagation)
- présence d'un bruit blanc additif gaussien

Le niveau du bruit relativement à celui du signal utile est décrit par le SNR (Signal to Noise Ratio) et exprimé en dB. On suppose que le SNR est connu lors de l'acquisition d'un signal. Lors de ce TP nous rencontrerons 4 niveaux de SNR: 30 dB (facile), 20 dB, 10 dB et 0 dB (en espérant qu'on puisse faire quelque chose de ces données).
Un de nos objectifs sera de qualifier la performance des algorithmes mis en place en fonction du SNR.

##  Chargement des données en Pytorch

Pour entrainer des réseaux de neurones profond sur nos données nous allons utiliser le framework Pytorch.
Une première étape va consister à transférer les données de numpy à PyTorch, cela passe par deux objets :

    - un Dataloader qui permet d'échantillonner le Dataset Pytorch dans les itérations de l'optimisation du réseau de neurones
    
Un dataset prend la forme
```python

        ...
```

Implémentez une classe Dataset pour le dataset considéré ici

````python
class MyDataset(torch.utils.data.Dataset):

  def __init__(self, path_to_data, transform = None):
    with h5py.File(path_to_data, 'r') as data:
      self.labels_id = np.array(data['labels'])
      self.signals = np.array(data['signaux'])
      self.snr = np.array(data['snr'])
    self.transform = transform   # Ajout d'une fonction transform, qui permet de faire la data augmentation

  def __len__(self):
    return len(self.labels_id)

  def __getitem__(self,i):
    signal, label = self.signals[i].T, self.labels_id[i]
    signal = torch.tensor(signal, dtype=torch.float32)  

    if self.transform:
      signal = self.transform(signal)
    return (signal, label)

````

Nous faisons ici de la data augmentation pour améliorer la performance de nos modèles 

Les valeurs que nous avons mis espour le bruit, le jitter, le shift, sont basés sur des fonctions que nous avons codées plus bas et qui sont sensées renvoyer les valeurs moyennes pour les signaux. Cependant, pour le shift et le jitter, les vameurs renvoyées étaient trop élevées et rendaient les signaux incompréhensibles. Nous avons donc seulement gardé le TimeShift et le bruit gaussien. La classe Compose permet d'appliquer plusieurs transformations à la suite

````python
## Pipeline de data augmentation

class FrequencyShift:
    def __init__(self, max_shift=0.02):
        self.max_shift = max_shift

    def __call__(self, signal):
        shift = np.random.uniform(-self.max_shift, self.max_shift)
        real = signal[0]  # Partie réelle
        imag = signal[1]  # Partie imaginaire
        
        # Créer le signal complexe
        complex_signal = torch.complex(real, imag)
        time = torch.arange(signal.shape[-1], dtype=torch.float32)  # Temps (index des échantillons)
        phase_shift = torch.exp(1j * 2 * np.pi * shift * time)  # Décalage de phase
        
        # Appliquer le décalage de fréquence au signal complexe
        shifted_signal = complex_signal * phase_shift
        
        # Retourner le signal sous forme réelle (partie réelle et imaginaire séparées)
        return torch.view_as_real(shifted_signal).float().T

class PhaseJitter:
    def __init__(self, max_jitter=0.1):
        self.max_jitter = max_jitter

    def __call__(self, signal):
        jitter = torch.rand(1) * 2 * torch.pi * self.max_jitter
        real = signal[0]
        imag = signal[1]
        complex_signal = torch.complex(real, imag)
        
        # Ajouter le jitter de phase
        phase_jitter = torch.exp(1j * jitter)  # appliquer l'exponentielle à la phase
        
        # Recalculer le signal avec le jitter
        jittered_signal = complex_signal * phase_jitter
        real_jittered = torch.real(jittered_signal)
        imag_jittered = torch.imag(jittered_signal)
        
        return torch.stack([real_jittered, imag_jittered], dim=0)

class AddGaussianNoise:
    def __init__(self, noise_level=0.1):
        self.noise_level = noise_level

    def __call__(self, signal):
        noise = torch.randn_like(signal) * self.noise_level
        return signal + noise

class TimeShift:
    def __init__(self, max_shift=1024):
        self.max_shift = max_shift

    def __call__(self, signal):
        # S'assurer que le signal est bien un tensor
        if not isinstance(signal, torch.Tensor):
            signal = torch.tensor(signal, dtype=torch.float32)

        # Appliquer le décalage temporel
        shift = np.random.randint(-self.max_shift, self.max_shift)
        shifted_signal = torch.roll(signal, shifts=shift, dims=-1)

        # Retourner un tensor (et non une instance de classe)
        return shifted_signal
    
class Compose:
    def __init__(self, transforms):
        self.transforms = transforms

    def __call__(self, signal):
        for transform in self.transforms:
            signal = transform(signal) 
        return signal
````

- charger les données
- les batcher

- logger l'évolution au fil des epochs  de la loss sur l'ensemble train et l'ensemble de validation et éventuellement de métriques complémentaires

````python
## Analyse des perturbations afin de bien calibrer les transformations de data augmentation
## Au final, les valeurs caractéristiques de frequency shift et de phase jitter sont sans doute surévaluées

def analyze_signal_perturbations(signals, sample_rate=1.0):
    # Liste pour stocker les statistiques
    noise_levels = []
    phase_variations = []
    frequency_shifts = []
    
    # Analyser chaque signal du dataset
    for signal, _ in signals:
        #print(signal)
        # Assurez-vous que le signal est sous forme complexe
        real_part = signal[0]  # Première ligne pour la partie réelle
        imag_part = signal[1]  # Deuxième ligne pour la partie imaginaire
        complex_signal = real_part + 1j * imag_part

        # 1. Estimation du bruit : calculer la variance après un lissage
        smooth_signal = np.convolve(np.abs(complex_signal), np.ones(10)/10, mode='same')
        noise = np.abs(complex_signal) - smooth_signal
        noise_variance = np.var(noise.detach().numpy())
        noise_levels.append(noise_variance)
        
        # 2. Déviation de fréquence : Analyse par transformée de Fourier
        N = len(complex_signal)
        yf = fft(complex_signal.detach().numpy())
        xf = fftfreq(N, 1 / sample_rate)
        peak_freq = xf[np.argmax(np.abs(yf))]
        frequency_shifts.append(abs(peak_freq))

        # 3. Variation de phase : Calcul de la dérivée de la phase
        phase = np.angle(complex_signal)
        phase_diff = np.diff(phase)
        phase_variation = np.std(phase_diff)
        phase_variations.append(phase_variation)
    
    # Résultats globaux sur l'ensemble des signaux
    results = {
        'average_noise_level': np.mean(noise_levels),
        'max_noise_level': np.max(noise_levels),
        'average_frequency_shift': np.mean(frequency_shifts),
        'max_frequency_shift': np.max(frequency_shifts),
        'average_phase_variation': np.mean(phase_variations),
        'max_phase_variation': np.max(phase_variations)
    }
    
    return results

print(analyze_signal_perturbations(dataset_test))

````

- Diminution du nombre de canaux, augmentation linéaire et non exponentielle afin qu'ils augmentent moins vite que le pooling

- Data augmentation (bruit + timeshift) = très efficace contre l'overfitting. On a une accuracy de 0.81 en 100 epoch, 0.824 en 200 epoch.

````python
class RNN2(nn.Module):
    def __init__(self):
        super(RNN2, self).__init__()
        self.rnn1 = nn.RNN(input_size=2048, hidden_size=8, num_layers=2, batch_first=True, dropout=0.5)
        self.fc = nn.Linear(in_features=8, out_features=6)

    def forward(self, x):
        # x est de taille [B, 2, 2048], donc pas besoin d'aplatir ici
        x, _ = self.rnn1(x)  # La sortie de rnn1 est de taille [B, 2, 32]
        x = x[:, -1, :]  # Récupère la dernière sortie temporelle pour chaque batch -> [B, 32]
        x = F.log_softmax(self.fc(x), dim=1)  # Applique la couche fully connected et softmax
        return x  # La sortie sera de forme [B, 6]
rnn=RNN2()
print(f'Il y a dans le modèle RNN2 {count_n_param(rnn)} paramètres')
````

````python
def evaluate_model(model, model_name): 
    all_acc=0
    snrs=dataset_test.snr
    snr_values=[0, 10, 20, 30]
    acc_snrs=[0, 0, 0, 0]
    compte_snrs=[0, 0, 0, 0]
    y_preds=[]
    y_reals=[]
    for i in range( ( dataset_test.__len__())): 
        x=torch.tensor(dataset_test.__getitem__(i)[0])
        x=x.reshape([1, x.shape[0], x.shape[1]])
        y_predicted, y_real = model(x),torch.tensor( dataset_test.__getitem__(i)[1])

        # y_predicted, y_real = model(torch.tensor(dataset_test.__getitem__(i)[0])),torch.tensor( dataset_test.__getitem__(i)[1])
        y_predicted=y_predicted.reshape(1, 6)
        y_real=torch.tensor(y_real.detach().numpy()).reshape(1)
        acc=accuracy(y_predicted,  y_real)  # l'accuracy prend un vecteur de probas et compare le np.argmax() de ce vecteur avec le vrai label. malynx
        all_acc+=acc
        snr=snrs[i]
        lab_snr=snr_values.index(snr)
        compte_snrs[lab_snr]+=1
        acc_snrs[lab_snr]+=acc
        y_real, y_predicted = (int(y_real.detach().numpy()[0]), int(np.argmax(y_predicted.detach().numpy())))
        y_preds.append(y_predicted)
        y_reals.append(y_real)

            
        

    disp=ConfusionMatrixDisplay( confusion_matrix(y_pred=y_preds, y_true=y_reals))
    disp.plot()
    plt.title('Confusion matrix for the model ' + model_name)
    plt.show()
    for i in range(len(acc_snrs)): 
        acc_snrs[i] = acc_snrs[i] / compte_snrs[i]
    plt.bar([0, 10, 20, 30], acc_snrs, edgecolor='black')
    plt.xlabel('SNR')
    plt.ylabel('Accuracy of the model')
    plt.title('Accuracy of the model ' + model_name+ ' with the SNR')
    plt.hlines(all_acc/dataset_test.__len__(), -10, 35,color = 'r' , linestyles='--', label='accuracy mean')
    plt.xlim(-2, 32)
    plt.legend()
    plt.show()

    print('accuracy of the model ' + model_name, all_acc/dataset_test.__len__())
````
