$\forall u=0,\dots,2L-1,\quad\phi_{L,k}[u]=w_{L}[u]\sqrt{\frac{2}{L}} \cos [ \frac{\pi}{L} \left(u+ \frac{L+1}{2}\right) (k+\frac{1}{2}) ]$
where $w_{L}$ is a modulating window given by
\begin{equation}
w_L[u] = \sin \left[{\frac {\pi }{2L}}\left(u+{\frac {1}{2}}\right)\right].
\end{equation}

```python
def w(L): 
    return np.array([np.sin(np.pi * (u + 0.5) / (2*L)) for u in range(2048)])

def phi(L, k) : 
    return w(L)*np.sqrt(2/L)*np.array([np.cos((np.pi/L)*(u+((L+1)/2))*(k+(1/2))) for u in range(2048)])
```

```python
scales = [32, 64, 128, 256, 512, 1024]
dictionary = build_mdct_dictionary(scales)
```

```python
def omp(signal, dictionary, num_atoms):

    residual = signal.copy()  # Initialize the residual
    indices = []  # Indices of selected atoms
    coeffs = []  # Coefficients for selected atoms
    norms=[]
    for _ in range(num_atoms):
        correlations = np.array([np.dot(residual, atom) for atom in dictionary])
        best_index = np.argmax(np.abs(correlations))
        indices.append(best_index)
        coeff=correlations[best_index]
        A = np.array([dictionary[i] for i in indices]).T
        x = np.linalg.lstsq(A, signal, rcond=None)[0]
        current_reconstruction=np.zeros(2048)
        for i in range(len(x)): 
            current_reconstruction+=x[i]*dictionary[indices[i]]
        coeffs=list(x)
        residual=signal-current_reconstruction
        norms.append(np.linalg.norm(residual))
    sparse_representation = np.zeros(dictionary.shape[0])
    for idx, coeff in zip(indices, coeffs):
        sparse_representation[idx] = coeff
    return sparse_representation, norms

signal_question_4
sparsity = 10  
sparse_representation , norms= omp(signal_question_4, dictionary, sparsity)
```
